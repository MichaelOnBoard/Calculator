```python
from tkinter import *

action = ""


def press(number):
    global action
    action += str(number)
    equation.set(action)


def equalpress():
    try:
        global action
        total = str(eval(action))
        equation.set(total)
        action = ""
    except Exception as e:
        equation.set("error")
        action = ""

def clear():
    global action
    action = ""
    equation.set("")

def create_button(gui, text, command, row, column):
    button = Button(gui, text=text, fg='black', bg='white', command=command, height=2, width=18)
    button.grid(row=row, column=column)

if __name__ == "__main__":
    gui = Tk()
    gui.configure(background="white")
    gui.title("Calculator")
    gui.geometry("500x450")

    equation = StringVar()
    
    action_field = Entry(gui, textvariable=equation)
    action_field.grid(columnspan=4, ipadx=70)

    buttons = [
        ('1', 2, 0), ('2', 2, 1), ('3', 2, 2),
        ('4', 3, 0), ('5', 3, 1), ('6', 3, 2),
        ('7', 4, 0), ('8', 4, 1), ('9', 4, 2),
        ('0', 5, 0), ('+', 2, 3), ('-', 3, 3),
        ('*', 4, 3), ('/', 5, 3), ('=', 5, 2),
        ('.', 5, 1), ('Clear', 6, 0),('Python Calculator', 6, 1)
    ]

    for (text, row, column) in buttons:
        if text == '=':
            create_button(gui, text, equalpress, row, column)
        elif text == 'Clear':
            create_button(gui, text, clear, row, column)
        else:
            create_button(gui, text, lambda x=text: press(x), row, column)

    gui.mainloop()
```
